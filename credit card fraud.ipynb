```python
# importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading dataset
df=pd.read_csv("C:/Users/dell/Desktop/creditcard.csv")
```

```python
# in the dataset 1st column is time elapsed after first transaction
# v1,....v28 are columns with some information about credit card transaction and converted to numerical values using PCA
# amount is transaction amount in US dollars
# class tells whether the transaction is fraudulent or not 0-> legit and 1 -> fraud 
```

```python
# this dataset is highly unbalanced
```

```python
# separating the data for analysis
legit = df[df.Class==0]
fraud = df[df.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
df.groupby("Class").mean()
```

```python
# under sampling
# build a sample dataset containing similar distribution of normal and fraud transactions
legit_sample = legit.sample(n=492)
```

```python
df1 = pd.concat([legit_sample, fraud], axis=0)     #axis =0 for row wise  and axis =1 for columns
```

```python
df1.head()
```

```python
df1.tail()
```

```python
df1["Class"].value_counts()
```

```python
df1.groupby("Class").mean()
```

```python
# splitting the data into featres and targets

X= df1.drop(columns="Class",axis=1)
Y= df1["Class"]
```

```python
print(X)
```

```python
print(Y)
```

```python
# split the data into training and testing data

X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size =0.2,stratify=Y,random_state=2)          # test_size =0.2 is that 20% for testing and 80 % for taining
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#modeltraining
# logistic regession(binary classification problem)
```

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

```python
#model evaluation
# accuracy score
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on Training Data : ",training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on Testing Data : ",testing_data_accuracy)
```
